#### Artificial intelligence

+ AI learning correctly

+ Put what AI learned into work


```python
# Creating

from sklearn.model_selection import train_test_split

y = table['Vendas']
x = table.drop('Vendas', axis=1)

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size =0.3, random_state=1)

# create this 4 variables
```

```python
from sklearn.linear_model import  LinearRegression
from sklearn.ensemble import  RandomForestRegressor



# Learning

# 2 models to compare which is better

model_linearRegression = LinearRegression()
model_RandonForestRegressor = RandomForestRegressor()

#     Trainning Linear Regression

model_linearRegression.fit(x_train, y_train) # fit == train the AI

#     Trainning RandomForest

model_RandonForestRegressor.fit(x_train, y_train)
```

#### AI results

Using R^2

```python
from sklearn import metrics

# create predict

predict_linearRegression = model_linearRegression.predict(x_test)
predict_ForestRegressor = model_RandonForestRegressor.predict(x_test)

# compare models

print(metrics.r2_score(y_test, predict_linearRegression))
print(f'{metrics.r2_score(y_test, predict_ForestRegressor):.2f}') 
```

```python
# Graphics view

table_assistant = pd.DataFrame()
table_assistant['y_test'] = y_test
table_assistant['Forest Regressor'] = predict_ForestRegressor
table_assistant['Linear Regression'] = predict_linearRegression

print('First graph')
plt.figure(figsize=(15,5)) # resize the graphics
sns.lineplot(data=table_assistant)
plt.show()
```

```python
# Who is most important to generate revenue

sns.barplot(x=x_train.columns, y=model_RandonForestRegressor.feature_importances_)
plt.show()
```
