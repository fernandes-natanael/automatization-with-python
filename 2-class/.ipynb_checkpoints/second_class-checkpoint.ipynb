```python
import pandas as pd

# Step 1

table = pd.read_csv('telecom_users.csv')
```

```python
# Step 2
#     Columns without name -> delete this
#     Columns with all values NaN
#     Columns 'Unnamed'without name
#     Columns with numeric values recognized like a text
#     Empty rows

print(table.info()) # show information from the table
display(table)
```

```python
# Step 3

table = table.drop("Unnamed: 0", axis = 1) # delete columns axis = 1 | delete row axis = 0 

table['TotalGasto'] = pd.to_numeric(table['TotalGasto'], errors='coerce') # convert the columns str to numeric
# coerce == leavy empty

table = table.dropna(how='all', axis=1 ) # delete all empty columns
# how='all' == all values needs to be empty
table = table.dropna(how='any', axis=0 ) # delete all row were has an empty value
# how='any' == at least one value needs to be empty

```

```python
# Step 4

display(table['Churn'].value_counts()) # quantity
display(table['Churn'].value_counts(normalize=True)) # percent
```
